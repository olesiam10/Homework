### 3. Find cities A and B where city A is warmer than city B in the summer, but also colder than city B in the winter, on average

```python
df['month'] = pd.DatetimeIndex(df['datetime']).month
df['month']
```

```python
season_bins = [1,3,6,9,12] #creating bins to group months by season to figure our the average temparatures for winter and summer
season_labels = ["winter", "spring", "summer", "fall"]

df["Seasons"] = pd.cut(df["month"], season_bins, labels=season_labels )

grouped_seasons = df.groupby("Seasons")

SeasonAverage = {"vanc_temp":grouped_seasons["Vancouver"].mean(),
                "san_diego_temp":grouped_seasons["San Diego"].mean(),
                 "las_vegas_temp":grouped_seasons["Las Vegas"].mean(),
                "portland_temp":grouped_seasons["Portland"].mean(),
                "los_ang_temp":grouped_seasons["Los Angeles"].mean()}

pd.DataFrame(SeasonAverage)
```

City A is Las Vegas and city B is San Diego. Las Vegas is warmer than San Diego in summer(303 degrees) and colder in winter(286). These 2 cities satisfy the assignment requirement.
