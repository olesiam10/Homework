```python
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
url = 'http://dndbeyond.com/classes'
browser.visit(url) 
```

```python
import pandas as pd

quote_dict = {'class': []}
for i in range(1,5):    
    soup = bs(browser.html, 'html.parser')
    all_divs = soup.find_all('div',class_='listing-card__content')
    for div in all_divs:
        try:
            class_text = div.find('h3',class_='listing-card__title').text
            class_text = class_text.strip()
            quote_dict["class"].append(class_text)

        except AttributeError as e:
            print(e)
            
print(quote_dict)
    
df = pd.DataFrame(quote_dict)
df
```

# 2. Now, alter 1 to click on each class name to go to the more detailed page, and pull out the first paragraph about each class. Store these in a dataframe. Your dataframe should have two columns.

```python
import pandas as pd
import time

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
url = 'http://dndbeyond.com/classes'
browser.visit(url)
tempurl = ''

paragraphs_dict = {'class': [],
             'paragraph': []}
for i in range(1,5):    
    soup = bs(browser.html, 'html.parser')
    all_divs = soup.find_all('div',class_='listing-card__content')
    for div in all_divs:
        try:
            class_text = div.find('h3',class_='listing-card__title').text
            class_text = class_text.strip()
            if class_text != "Artificer" and class_text != "Blood Hunter":
                class_url = class_text.replace(' ', '-').lower()
                tempurl = url + "/" + class_url
                browser.visit(tempurl)
                
                #sleep timer here in case you have to say you're a human and press the button
                time.sleep(10)

                soup = bs(browser.html, 'html.parser')
                par_divs = soup.find_all('p',class_='Core-Styles_Core-Body')

                paragraph_text = par_divs[0].text
                print(paragraph_text)

                paragraphs_dict["class"].append(class_text)
                paragraphs_dict["paragraph"].append(paragraph_text)


        except AttributeError as e:
            print(e)
            
print(paragraphs_dict)
    
df = pd.DataFrame(paragraphs_dict)
df
```
